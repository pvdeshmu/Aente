```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#This the file with all professions: Healthcare and non healthcare
# Load the data from the Excel file
file_path = 'all_data_M_2023_cleaned.xlsx'
data = pd.read_excel(file_path)
```

```python
print(data.head())
```

```python
print(data.columns)
```

```python

# Display unique values from the "Industry" column to identify all medical professions
unique_industries = data['OCC_TITLE'].unique()

# Print the unique industries to identify medical professions
for industry in unique_industries:
    print(industry)
```

```python
# Convert the wage column to numeric, forcing any errors to be NaN
data['A_MEDIAN'] = pd.to_numeric(data['A_MEDIAN'], errors='coerce')

# Sort the data by A_MEDIAN and select the top 10 occupations
top_10_occupations = data[['OCC_TITLE', 'A_MEDIAN']].sort_values(by='A_MEDIAN', ascending=False).head(10)

# Plotting the data
plt.figure(figsize=(14, 8))
bars = plt.barh(top_10_occupations['OCC_TITLE'], top_10_occupations['A_MEDIAN'], color='blue')

# Add labels to each bar
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, 
             round(bar.get_width(), 2), 
             va='center', ha='left', fontsize=10, color='black')

# Set the title and labels
plt.title('Top 10 Occupations by Median Annual Wage')
plt.xlabel('Median Annual Wage ($)')
plt.ylabel('Occupation')
plt.tight_layout()

# Save the plot as a PNG image
plt.savefig('top_10_occupations_by_median_wage.png')

# Display the plot
plt.show()
```

```python
# Convert the wage column to numeric, forcing any errors to be NaN
data['A_MEDIAN'] = pd.to_numeric(data['A_MEDIAN'], errors='coerce')

# Group by occupation and calculate the average of the median annual wage
average_median_wages = data.groupby('OCC_TITLE')['A_MEDIAN'].mean().sort_values(ascending=False)

# Select the top 10 occupations by average of median wage
top_10_occupations = average_median_wages.head(20)

# Sort the occupations from high to low
top_10_occupations = top_10_occupations.sort_values(ascending=True)

# Plotting the data
plt.figure(figsize=(14, 8))
bars = plt.barh(top_10_occupations.index, top_10_occupations.values, color='blue')

# Add labels to each bar
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, 
             f'${round(bar.get_width(), 2):,.2f}', 
             va='center', ha='left', fontsize=10, color='black')

# Set the title and labels
plt.title('Top 10 Occupations by Average of Median Annual Wages')
plt.xlabel('Median Wage ($)')
plt.ylabel('Industry and Profession')
plt.tight_layout()

# Adjust the x-axis limits for better visualization
plt.xlim(0, max(top_10_occupations.values) * 1.1)

# Save the plot as a PNG image
plt.savefig('top_20_occupations_by_average_median_wage.png')

# Display the plot
plt.show()
```

```python
# Convert necessary columns to numeric, forcing any errors to be NaN
data['A_MEDIAN'] = pd.to_numeric(data['A_MEDIAN'], errors='coerce')
data['TOT_EMP'] = pd.to_numeric(data['TOT_EMP'], errors='coerce')

# Define industry titles for major sectors
'''industries_to_analyze = {
    'Healthcare (Providers)': ["Family Medicine Physicians", "General Internal Medicine Physicians", "Pediatricians", 
                             "Physicians and Surgeons", "Surgeons", "Nurse Anesthetists", "Nurse Practitioners", "Nurse Midwives"],
    'Technology': ['Software Developers', 'Information Security Analysts', 'Computer Systems Analysts'],
    'Legal': ['Lawyers', 'Judges, Magistrate Judges, and Magistrates'],
    'Financial Services': ['Financial Analysts', 'Personal Financial Advisors', 'Securities, Commodities, and Financial Services Sales Agents'],
    'Pharmaceuticals': ['Pharmacists', 'Medical Scientists, Except Epidemiologists'],
    'Engineering and Architecture': ['Architects, Except Landscape and Naval', 'Civil Engineers', 'Mechanical Engineers']
}'''


        
'''"Physician Assistants", "Nurse Midwives", "Nurse Practitioners", "Veterinarians", "Physicians",  "Cardiologists", "Dermatologists",
        "Emergency Medicine Physicians", "Family Medicine Physicians", "Neurologists",
        "Obstetricians and Gynecologists"'''

industries_to_analyze = {
    'Healthcare (Providers)': [ "Dentists",  "Optometrists",
         "Physician Assistants", "Podiatrists", "Occupational Therapists", "Nurse Midwives",
        "Nurse Practitioners", "Physicians", "Anesthesiologists", "Cardiologists", "Dermatologists",
        "Emergency Medicine Physicians", "Family Medicine Physicians", "General Internal Medicine Physicians", "Neurologists",
        "Obstetricians and Gynecologists", "Pediatricians, General", "Psychiatrists",
        "Radiologists", "Physicians, All Other", "Surgeons", "Ophthalmologists, Except Pediatric", "Orthopedic Surgeons, Except Pediatric",
        "Pediatric Surgeons"],
    'Technology': [ "Database and Network Administrators and Architects", "Database Architects", "Network and Computer Systems Administrators",
        "Software and Web Developers, Programmers, and Testers", "Computer Programmers", "Software Developers", "Software Quality Assurance Analysts and Testers",
        "Web Developers"],
    'Legal': [ "Lawyers and Judicial Law Clerks", "Lawyers"],
    'Financial Services': ["Financial Managers", "Financial Specialists", "Budget Analysts", "Credit Analysts", "Financial Analysts and Advisors",
        "Financial and Investment Analysts", "Personal Financial Advisors", "Financial Risk Specialists", "Financial Examiners", "Credit Counselors", "Loan Officers", "Tax Examiners, Collectors and Preparers, and Revenue Agents",
        "Tax Preparers", "Miscellaneous Financial Specialists", "Financial Specialists, All Other",
        "Securities, Commodities, and Financial Services Sales Agents"],
    'Pharmaceuticals': [ "Pharmacists", "Chemical Technicians",
        "Nuclear Technicians", "Clinical Laboratory Technologists and Technicians"],
    'Engineering and Architecture': [ 'Civil Engineers', 'Mechanical Engineers']
}


# Assume the total revenue for each industry (in trillions for demonstration purposes)
industry_revenue = {
    'Healthcare (Providers)': 40_000_000_000_000,  
    'Technology': 12_000_000_000_000,  
    'Legal': 4_000_000_000_000,       
    'Financial Services': 14_000_000_000_000,  
    'Pharmaceuticals': 4_000_000_000_000,     
    'Engineering and Architecture': 2_600_000_000_000  
}






# Calculate the percentage of revenue going to high-skilled labor for each industry
industry_percentages = {}

for industry, professions in industries_to_analyze.items():
    # Filter data for the current industry professions
    industry_data = data[data['OCC_TITLE'].isin(professions)]
    
    # Calculate the average median wage
    average_wage = industry_data['A_MEDIAN'].mean()
    
    # Calculate the total number of employees
    total_employees = industry_data['TOT_EMP'].sum()
    
    # Calculate total compensation for the industry
    total_compensation = average_wage * total_employees
    
    # Calculate the percentage of revenue going to high-skilled labor
    revenue_percentage = (total_compensation / industry_revenue[industry]) * 100
    
    # Store the result
    industry_percentages[industry] = revenue_percentage

# Convert the dictionary to a DataFrame for plotting
industry_df = pd.DataFrame.from_dict(industry_percentages, orient='index', columns=['Percentage']).sort_values(by='Percentage')

# Plotting the data
plt.figure(figsize=(12, 6))
bars = plt.barh(industry_df.index, industry_df['Percentage'], color='lightBlue')

# Add labels to each bar
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, 
             f'{round(bar.get_width(), 2)}%', 
             va='center', ha='left', fontsize=10, color='black')

# Set the title and labels
plt.title('Percentage of Industry Revenue Going to High-Skilled Labor')
plt.xlabel('Percentage of Revenue (%)')
plt.ylabel('Industry')
plt.tight_layout()

# Save the plot as a PNG image
plt.savefig('industry_revenue_comparison.png')

# Display the plot
plt.show()
```
